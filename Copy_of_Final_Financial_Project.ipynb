```python
# Importing financial data library

import yfinance as yf
```

```python
# Importing Libraries

import matplotlib.pyplot as pyplt
import datetime as date_obj
import pandas_datareader.data as reader_obj
import pandas as panda
import numpy as np
```

```python
# Loading 10 risky assets
# we are taking 10 risky assets' closing price for 3 months starting from 1 Oct, 2022 to 31 Dec, 2022

starter=date_obj.datetime(2022,12,1)
end=date_obj.datetime(2023,2,28)

Companies=['META','AMZN','GOOGL','JPM','TSLA','BABA','SWN','ZM','PYPL','NFLX']
tickers1 =['GSPC']

d_frame=panda.DataFrame()
for t in Companies:
    d_frame2=yf.download(t, start='2022-11-01',end='2023-01-31')['Close']
    d_frame[t]=d_frame2

#Printing Data
print("Closing Price of 10 Risky Assets : ")
print(d_frame)
```

```python
# Calculating Returns of each asset

return1=np.log(d_frame/d_frame.shift(1))
print("Here are the Returns : ")
print(return1)

return1.dropna(inplace=True)

print(return1)
```

```python
#Calculating Expected Return(mean), Variance and Standard Deviation of these returns.

Er=np.array(return1.describe().loc['mean'])
std=np.array(return1.describe().loc['std'])
var=std**2
varcovar1=return1.cov()

print("Expected Return  :  ",Er)

print("Standard Deviation   :  ",std)

print("Variance  :  ",var)

print("Covariance Matrix : ")
print(varcovar1)
```

**Calculation for Efficient Frontier Curve**

```python
# Efficient Portfolio A

weightA=np.linalg.inv(varcovar1) @ (Er) / np.sum(np.linalg.inv(varcovar1) @ (Er))
ErA=weightA@Er
VarA=weightA@varcovar1@weightA.T
stdA=np.sqrt(VarA)
print("Expected Return of portfolio A :  ",ErA)

print("Standard Deviation of portfolio A  :  ",stdA)

print("Variance of portfolio A :  ",VarA)

print("Weights of portfolio A  :")
print(weightA)

print("Ones Matrix :")
O_mat=np.ones((varcovar1.shape[0]))
print(O_mat)

# Efficient Portfolio B

c2=0.039

weightB=np.linalg.inv(varcovar1) @ (Er-c2) / np.sum(np.linalg.inv(varcovar1) @ (Er-c2))
ErB=weightB@Er
VarB=weightB@varcovar1@weightB.T
stdB=np.sqrt(VarB)

covAB=weightA@varcovar1@weightB.T

#Calculating weigts minimum

GMVP=O_mat @ np.linalg.inv(varcovar1) / (O_mat @ np.linalg.inv(varcovar1) @ O_mat.T)
GMVP
ErG=GMVP@Er   # Mu minimum
stdG=np.sqrt(GMVP@varcovar1@GMVP.T)
VarianceG =stdG**2    # Variance minimum

wa=np.linspace(-1.5,4.5,num=50)
wb=1-wa
ErC = wa*ErA + wb*ErB
stdC = np.sqrt((wa*stdA)**2 + (wb*stdB)**2 + 2*wa*wb*covAB)
VarianceC = stdC**2

print("Portfolio means :", ErC)
print("Portfolio variance :", VarianceC)
```

```python
# Plotting Efficient Frontier Curve

print("Efficient Frontier Curve")

#pyplt.scatter(VarianceC,ErC,color='m')
pyplt.grid(True)
pyplt.scatter(VarianceG,ErG,color='k',linewidth = 2.5)
pyplt.plot(VarianceC,ErC,'m',linewidth = 2.5)
pyplt.xlabel("Portfolio Variance ")
pyplt.ylabel("Portfolio Expected Return ")
pyplt.title("Efficient Frontier Curve")
pyplt.show()
```

```python


data = {'Returns':ErA, 'Volatility':stdA}
data1 = {'Returns':ErC, 'Volatility':stdC}

portfolios = panda.DataFrame(data1)

print(portfolios)
```

**Calculation for CAPM**

```python
# Taking risk free expected return
mu_rf = 0.00050
((portfolios['Returns']-mu_rf)/portfolios['Volatility']).idxmax()

optimal_risky_port = portfolios.iloc[((portfolios['Returns']-mu_rf)/portfolios['Volatility']).idxmax()]

cal_x = []
cal_y = []
utility = []
a = 10

for er in np.linspace(mu_rf, max(ErC), 50):
    sd = (er - mu_rf)/((optimal_risky_port[0]-mu_rf)/optimal_risky_port[1])
    u = er - .5*a*(sd**2)
    cal_x.append(sd)
    cal_y.append(er)
    utility.append(u)

data2 = {'utility':utility, 'cal_y':cal_y, 'cal_x':cal_x}
cal = panda.DataFrame(data2)
cal.head()
```

```python
# Plotting CAPM with straight line tangent to Markowitz Bullet

print(" CAPM ")
#pyplt.scatter(stdC,ErC)
pyplt.grid(True)
pyplt.scatter(stdG,ErG, linewidth = 2.5)
pyplt.plot(stdC,ErC,'b',linewidth = 2.5)
pyplt.plot(cal_x, cal_y,'r')
pyplt.xlabel("Portfolio Standard Deviation ")
pyplt.ylabel("Portfolio Expected Return ")
pyplt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='k',linewidth = 2.5)
pyplt.title("CAPM with markowitz bullet and tangent line")
pyplt.show()
```
